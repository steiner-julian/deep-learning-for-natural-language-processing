```python
import numpy as np
from matplotlib import pyplot as plt
```

## Functions

Define functions f and f'.

```python
def f(x):
    return x**2
```

```python
def f_derivative(x):
    return 2*x
```

## Gradient based Optimization

```python
def gradient_optimization(
    function,
    derivative,
    learning_rate=0.05,
    epochs=20,
    x_i=4
):
    x_values = []
    y_values = []

    for _ in range(epochs):
        u = derivative(x_i)
        x_i = x_i - learning_rate * u
        x_values.append(x_i)
        y_values.append(function(x_i))
    
    return x_i, x_values, y_values
```

```python
minimun, x_values, y_values = gradient_optimization(
    function=f,
    derivative=f_derivative,
    learning_rate=0.05,
    epochs=100
)
```

```python
minimun
```

## Visualizations

```python
x = np.linspace(-10, 10, 10)
plt.plot(x, f(x), color='red')
plt.scatter(x_values, y_values)
plt.show()
```
