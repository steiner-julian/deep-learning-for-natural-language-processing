```python
import numpy as np
from matplotlib import pyplot as plt
```

## Gradient based Optimization

```python
def gradient_optimization(
    f,
    f_derivative,
    learning_rate=0.05,
    epochs=20,
    x_i=4
):
    x_values = []
    y_values = []

    for _ in range(epochs):
        u = f_derivative(x_i)
        x_i = x_i - learning_rate * u
        x_values.append(x_i)
        y_values.append(f(x_i))
    
    return x_i, x_values, y_values
```

## Visualizations

```python
def plot_gradient_optimization(
    f,
    f_derivative,
    learning_rate,
    epochs,
    x_i=4,
):
    minimun, x_values, y_values = gradient_optimization(
        f=f,
        f_derivative=f_derivative,
        learning_rate=learning_rate,
        epochs=epochs,
        x_i=x_i
    )

    print(f'Minimum: {minimun}')

    x = np.linspace(-10, 10, 50)
    plt.plot(x, f(x), color='red')
    plt.plot(x_values, y_values, color='green')
    plt.scatter(x_values, y_values)
    plt.show()
```

## Examples

### Functions

```python
def f(x):
    return x**2

def f_derivative(x):
    return 2*x
```

```python
plot_gradient_optimization(
    f=f,
    f_derivative=f_derivative,
    learning_rate=0.1,
    epochs=20
)
```

```python
def f(x):
    return x**3 - 4*x

def f_derivative(x):
    return 3*x**2 - 4
```

```python
plot_gradient_optimization(
    f=f,
    f_derivative=f_derivative,
    learning_rate=0.1,
    epochs=10
)
```

```python
def f(x):
    return 3 - (np.sin(2 * x) / x)

def f_derivative(x):
    return (np.sin(2*x) - 2 * x * np.cos(2*x) / x**2)
```

```python
plot_gradient_optimization(
    f=f,
    f_derivative=f_derivative,
    learning_rate=0.1,
    epochs=1000,
    x_i=1.5
)
```
