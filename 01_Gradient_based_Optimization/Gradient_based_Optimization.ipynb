## Functions

Define functions f and f'.

```python
def f(x):
    return x**3
```

```python
def f_derivative(x):
    return 3*(x**2)
```

## Gradient based Optimization

```python
def gradient_optimization(
    function,
    derivative,
    learning_rate=0.05,
    epochs=20,
    x_i=4
):
    x_values = []
    y_values = []

    for _ in range(epochs):
        u = derivative(x_i)
        x_i = x_i - learning_rate * u
        x_values.append(x_i)
        y_values.append(function(x_i))
    
    return x_i, x_values, y_values
```
